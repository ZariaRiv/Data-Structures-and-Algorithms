```python
from math import sqrt

class Vector ():
    def __init__(self, dimension, numbers ):
        self.dimension=dimension
        self.numbers=numbers
        sumofsquares=0
        for element in self.numbers:
            to_be_added=element*element
            sumofsquares+=(to_be_added)    
        self.length=sqrt(sumofsquares)
        
    def __repr__( self ):
        return "{}".format( self.numbers )
        
    def __ge__( self, another ): #>=
        another=Vector(another[0],[1])
        return  self.length>=another.length
    
    def __ne__( self, another ):     #!=
        return self.numbers!=another.numbers
    
    def __lt__( self, another  ):      #<
        return  self.length<another.length
    
    def __le__( self, another  ):       #<=
        
        return  self.length<=another.length
    
    def __gt__( self, another  ):      #>
        
        return  self.length>another.length
    
    def __ge__( self, another  ):            #>=
        
        return  self.length<=another.length
    
    
    def __add__( self, another  ):
        newvector=Vector(self.dimension, [])
        if self.dimension != another.dimension:
            return " You can't add vectors with different number of dimensions"
        else:
            i=0
            while i < (self.dimension):
                newvector.numbers.append( self.numbers[i]+another.numbers[i])
                i+=1
            return newvector
                
    
    def __sub__( self, another  ):
        newvector=Vector(self.dimension, [])
        if self.dimension != another.dimension:
            return " You can't substract vectors with different number of dimensions"
        else:
            i=0
            while i < (self.dimension):
                newvector.numbers.append( self.numbers[i]-another.numbers[i])
                i+=1
            return newvector
        
    def __mul__( self, scalar  ):
        newvector=Vector(self.dimension, [])
        i=0
        while i < (self.dimension):
            newvector.numbers.append( self.numbers[i]*scalar)
            i+=1
        return newvector
        
        
    def __rmul__( self, scalar ):
        return self.__mul__(scalar)
        
    def __neg__( self  ):
        newvector=Vector(self.dimension, [])
        i=0
        while i < (self.dimension):
            newvector.numbers.append( -self.numbers[i])
            i+=1
        return newvector
    
    def __abs__( self  ):
        newvector=Vector(self.dimension, [])
        i=0
        while i < (self.dimension):
            newvector.numbers.append( abs(self.numbers[i]))
            i+=1
        return newvector
    
    def __len__( self  ):
        return  self.dimension
    
    def __getitem__( self, n ):
        return self.numbers[n]
        
    def __setitem__( self, n, value ):
        self.numbers[n] = value
        
    def __contains__( self, number):
        return number in self.numbers
        
#vektor= Vector(4, [1,2,3,0])
#drugivektor= Vector(4, [1,2,3,0])
#print(vektor.length)


#Suppose that you use 3-dimensional vectors, with v1=(1,2,3)and v2=(2,1,4)

v1=Vector(3,[1,2,3])
v2=Vector(3,[2,1,4])

print( v1 != v2)
print( v1 < v2)
print( 3 in v1)
print(5 not in v1)

print( v1 + v2 )  # (3,3,7)
print( v1-v2 )   # (-1,1,-1)
print( v1*2 )     # (2,4,6)
print( 2*v1 )     # (2,4,6)
print(-v1)        # (-1,-2,-3)
abs(-v1)          # (1,2,3)
print(len(v1)  )
# 3
print()
print(v1)
v1[2] = 3
print(v1)
v1[-1]=7          # v1 = (1,2,7)
print(v1)
```
