```python
from copy import deepcopy
state=["c","c","c","","b","b","b"]
#     [ 0, 1 ,  2,  3, 4,  5,  6 ]
def possible_moves(state):
    x=0
    moves=[]
    while x<len(state):
        new_state=deepcopy(state)
        if state[x]=="b":
            if x==1 and state[0]=="":
                state[0]="b"
                state[1]=""
            elif x==0:
                continue
            else:
                if state[x-2]==""and state[x-1]=="c":
                    new_state[x-2]="b"   
                    new_state[x]=""
                    #print(1)
                    moves.append(new_state)
                elif state[x-1]=="":
                    new_state[x-1]="b" 
                    new_state[x]=""
                    #print(2)
                    moves.append(new_state)
        elif state[x]=="c":
            maxi=len(state)
            if x==(maxi-2) and state[maxi-1]=="":
                state[maxi-1]="b"
                state[maxi-2]=""
            elif x==(maxi-1):
                continue
            else:
                if state[x+2]=="" and state[x+1]=="b":
                    new_state[x+2]="c"
                    new_state[x]=""
                    #print(3)
                    moves.append(new_state)
                elif state[x+1]=="":
                    new_state[x+1]="c"
                    new_state[x]=""
                    #print(4)
                    moves.append(new_state)
        x+=1
    return(moves)
            
def goats( state, endstate, solution ):
    if state == endstate:
        solution=state
        return solution
    else:
        options=possible_moves(state)
        if options==[]:
            return False
        for move in options:
            new_state=move
            goats(new_state,endstate, solution)
            
            
            
goats(["c","c","","b","b"], ["b","b","","c","c"], [])
```
