```python
from random import randint 
how_many_want=int(input("How long you want your word to be?\n (remember, how many letters there is, that many chances to guess the word)"))+1
possible_words=[]
number_of_words=0
choosen_letter=0
word=[]

with open( "wordsEn.txt" )as fp:
    while True:
        while True:
            buffer = fp.readline()
            if buffer == "":
                break
            how_many= len(buffer)
            if how_many == how_many_want:
                possible_words.append(buffer)
                number_of_words+=1
        
        #print(possible_words)
        number_of_words-=1
        break

        
        
which_word=randint(0,number_of_words)
choosen_word=possible_words[which_word]
choosen_word=choosen_word.upper().strip()
#print(choosen_word)



how_many_want-=2
choosen_letter=randint(0,how_many_want)
#print(choosen_letter)


for char in choosen_word:
    word.append(char)

#print(word)
            
i=0
how_many_want+=1
while i<how_many_want:
    if i == choosen_letter:
        print( word[i], end="" )
    else:
        print("-", end="")
    i+=1
print("\n")

    
# now user quess
chances=0
while chances<=how_many_want:
    guess=input("Now guess the word :) ")
    lenghtcheck=len(guess)
    if lenghtcheck != how_many_want:
        print("Please try to guess the word indicating correct number of letters..")
        continue
    else:
        guessing=[]
        guessed=[]
        for char in guess.upper():
            guessing.append(char)
            guessed.append("_")
    
        j=0
        while j<how_many_want:
            if word[j] == guessing[j]:
                guessed[j]=word[j]
                print( guessed[j], end="" )
            else:
                print("-", end="")
            j+=1
        chances+=1
        print("\n","remained chances:", how_many_want - chances, "\n" )
        
        if guessed == word:
            print("YAAS, that's the word!")
            break
print("The word was: ",choosen_word )
```
