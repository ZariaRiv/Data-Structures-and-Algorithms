```python
class Person:
    def __init__( self, age=0, height=0 ):
        self.age = abs(age)
        self.height = abs(height)
    def __repr__( self ):
        return "({}, {})".format( self.age, self.height )
    
class City:
    def __init__( self, name ):
        self.name = name
        self.population=[]
    def __repr__( self ):
        return "({}, {})".format( self.name, self.population )
    
    def populate(self, someone):
        self.population.append( someone )
            
    def avarage_height( self):
        lenght=len(self.population)
        suma=0
        for element in self.population:
            suma+=element.height
        return suma/lenght
    
    def highest_age( self):
        age_list=self.population
        #age_list.sort(key=self.age,reverse=True)
        self.highest_age=0
        for element in age_list:
            if element.age > self.highest_age:
                self.highest_age=element.age
        return self.highest_age
           
    def auto_populate(self):
        for i in range (0,101):
            from random import randint
            allowed_age=randint(18,100)
            allowed_height=randint(150,200)
            person = Person(allowed_age,allowed_height)
            self.populate(person)
        return self.population


class Country:
    def __init__( self, name=""):
        self.name=name
        self.population_country=[]
    def __repr__( self ):
        return "({}, {})".format( self.name,self.population_country )
    
    def populate(self, city):
        self.population_country.append( city )
    
    
    
#instanciate two cities ( TIlburg and Eindhoven) and one country the Netherlands    
Tilburg=City("Tilburg")
Eindhoven=City("Eindhoven")
NL=Country("The_Netherlands")


#populate Tilburg and Eindhoven with 5 persons each 
for i in range (0,10):
    from random import randint
    allowed_age=randint(18,100)
    allowed_height=randint(150,200)
    person = Person(allowed_age,allowed_height)
    if i%2==0:
        Tilburg.populate(person)
    else:
        Eindhoven.populate(person)
        

        
#look at what we have so far     
print(Tilburg, Eindhoven, NL)

#average height of Eindhoven
print(Eindhoven.avarage_height())

#highest age in Tilburg
print(Tilburg.highest_age())
        
# populate Tilburg with 100 people
Tilburg.auto_populate()

#put Tilburg in the Netherlands
NL.populate(Tilburg)


#see the country
#print(NL)
```
